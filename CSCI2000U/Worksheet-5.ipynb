```python
import seaborn as sns
import matplotlib as mpl
d_diam = sns.load_dataset('diamonds')
d_ice = sns.load_dataset('seaice')
```

```python
# 1.
p = sns.scatterplot(data=d_diam, x = 'carat', y = 'price', hue='cut')
# 2.
p = p.set(title="The price of diamonds", xlabel = "Weight (carats)", ylabel = "Price ($)")
```

```python
# 3.
p = mpl.pyplot.plot(d_ice.Extent, color='black', linestyle ='dashed')

# 4.
mpl.pyplot.xticks(range(0, 12001, 3000), labels=['1980', '1992', '2000', '2008', '2016'])

# 5.
mpl.pyplot.xlabel('Year')
mpl.pyplot.ylabel('Thickness (m)')
mpl.pyplot.suptitle('Change in Artic sea ice thickness over time')
pass
```
